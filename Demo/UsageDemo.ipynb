Description:
This demo calls the FRCfuncs library and runs demos on two .raw images (taken from Original release by Castello et al.) and two .txt files generated using Ani (developen by Suren Felekyan) based on lifetime from a single image.

Requirements:
Have the FRCfuncs.py file located in a dir pointed to by PYTHONPATH.
To set PYTHONPATH:
1. edit environmental variables
2. global variables new/ edit PYTHONPATH
3. add location of FRCfunc.py to PYTHONPATH separated by semicolon
4. close and restart all python processes. If that does not work, restart pc.
5. Alternatively, you can place FRCfunc.py in the folder your notebook is located in. This is not recommended as you'll create multiple copies and this will aggrevate version control.

Usage:
1. Read in image with corresponding routine
2. set pixelsize

Optionally:
3. set title used in plotting
4. set the theta in case you want to separate resolution in x and y.
5. adjust meanfilterwidth (must be int)

```python
from FRCfuncs import *
```

### analyse two .txt images generated at HHU

```python
pixelSize = 25 #in nm
title = 'Beads recorded at HHU'
fileA = r"demoImages\PQSpcm_2019-02-01_15-59-23-chan0-25.txt"
fileB = r"demoImages\PQSpcm_2019-02-01_15-59-23-chan25-200.txt"

imA, imB = readtxt(fileA, fileB)

__, BeadsFRC, BeadsthreeSigma, BeadsfiveSigma, __, Beadsres = FRCAnalysis(
    imA, imB, pixelSize, theta = 0, title = title)


print("for file Beads:\n " +
     "resolution threshold for 1/7, three sigma and five sigma is \n" + 
     str(CLSM02res[:,0]) + "\n\n")
```
