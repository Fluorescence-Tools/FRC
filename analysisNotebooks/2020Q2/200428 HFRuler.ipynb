```python
maxval = 60

outdir = os.path.join(rootdir, 'analysis_NV', 'chifit')
data = HF28Apr_an
datas = df.filterFRETind(data, 'stoichiometry', 0.2, 0.8)

datas = df.filterFRETind(datas, 'NG', 50, 1000)
datas = df.filterFRETind(datas, 'NY', 100, 1000)

fstats = df.genStats(datas)
#fstats = df.filterStats(fstats, 'sigmaY', 1.5, 2.5)
#fstats = df.filterStats(fstats, 'sigmaG', 0, 3.3)
plt.plot(fstats['distx'], fstats['disty'], 'o', alpha = 0.5)
plt.xlim(-30, 30)
plt.ylim(-30, 30)
plt.xlabel('dist x (nm)')
plt.ylabel('dist y (nm)')
plt.show()


dist = np.array((fstats['distx'], fstats['disty'])).transpose()
#dist = df.filterVec(dist, maxval = maxval, center = [5,0])
dist = df.filterVec(dist, maxval = maxval)
distnorm = np.linalg.norm(dist, axis = 1)
outfile = os.path.join(outdir, identifier + r'_distdict.txt')
distDict = {}
distDict ['distx'] = dist[:,0]
distDict ['disty'] = dist[:,1]
distDict['distnorm'] = distnorm
df.saveDict(distDict, outfile)


modelout = os.path.join(outdir, identifier + r'_chiFit.txt')
plotout = os.path.join(outdir, identifier + r'_subpopulation.png')

TACout = os.path.join(outdir, 'filterTAC\\' + identifier + '_cut.txt')
df.subensembleTAC(datas, outfile = TACout)

dist = distnorm
maxbin = maxval
binwidth = 1
counts, bin_edges, _ = plt.hist(dist, bins = np.arange(0, maxbin, binwidth))
plt.clf()
Nbins = bin_edges.shape[0] - 1
bins = np.zeros(Nbins)
for i in range(Nbins):
    bins[i] = (bin_edges[i] + bin_edges[i + 1]) / 2


fitres = df.whichChiIsBest(bins, counts, verbose = True)
print(fit_report(fitres))
df.plotdistr(distnorm, bins, fit = fitres , title = identifier, 
                           modelout = modelout, plotout = plotout)
p = df.genPeakEst(1, counts, bins)
p['bg'].set(vary = True, value = 0)
p['mu0'].set(vary = True, value = 0)
#p['mu0'].set(vary = False, value = 7.1)
#p['sig0'].set(vary = False, value = 4.6)
fitres2, AIC, *_ = df.fitNChidistr(p, bins, counts)

modelout = os.path.join(outdir, identifier + r'_singlechi_mu0.txt')
plotout = os.path.join(outdir, identifier + r'_singlechi_mu0.png')
df.plotdistr(distnorm, bins, fit = fitres2 , title = identifier + '_singlechi_mu0', 
                           modelout = modelout, plotout = plotout)
print(fit_report(fitres2))
print('AIC for mu fixed to 0 os %.1f' % AIC)
p = df.genPeakEst(2, counts, bins)
p['bg'].set(vary = True, value = 0)
p['mu0'].set(vary = True, value = 0)
p['sig0'].set(vary = True, value = 3.5)
#p['A0'].set(vary = True, value = 150)
#p['mu1'].set(vary = False, value = 7)
#p['sig1'].set(vary = False, value = 2)
#p['mu2'].set(vary = True, value = 5)
#p['sig2'].set(vary = True, value = 12)
fitres3, AIC, *_ = df.fitNChidistr(p, bins, counts)

modelout = os.path.join(outdir, identifier + r'_2chiFit.txt')
plotout = os.path.join(outdir, identifier + r'_2chiFit.png')
df.plotdistr(distnorm, bins, fit = fitres3 , title = identifier + '  - double chi', 
                           modelout = modelout, plotout = plotout)
print(fit_report(fitres3))
print('AIC for 2chi distributions is %.1f' % AIC)
outname = os.path.join (outdir, 'single_Chi_components.csv')
df.exportChiComponents(outname, bins, fitres3.params)

p = df.genPeakEst(2, counts, bins)
p['bg'].set(vary = True, value = 0)
p['mu0'].set(vary = False, value = 0)
p['sig0'].set(vary = True, value = 3.5)
#p['A0'].set(vary = True, value = 150)
#p['mu1'].set(vary = False, value = 7)
#p['sig1'].set(vary = False, value = 2)
#p['mu2'].set(vary = True, value = 5)
#p['sig2'].set(vary = True, value = 12)
fitres4, AIC, *_ = df.fitNChidistr(p, bins, counts)

modelout = os.path.join(outdir, identifier + r'_2chiFit_mu0.txt')
plotout = os.path.join(outdir, identifier + r'_2chiFit_mu0.png')
df.plotdistr(distnorm, bins, fit = fitres4, title = identifier + '  - double chi mu0 pop', 
                           modelout = modelout, plotout = plotout)
p = df.genPeakEst(1, counts, bins)
p['bg'].set(vary = False, value = 0)
p['mu0'].set(vary = False, value = 7)
p['sig0'].set(vary = False, value = 3.5)
p['A0'].set(vary= False, value = 300)
dummy_fit = fitres4
dummy_fit.params = p
modelout = os.path.join(outdir, identifier + r'_expected_distribution.txt')
plotout = os.path.join(outdir, identifier + r'_expected_distribution.png')
df.plotdistr(distnorm, bins, dummy_fit, title = identifier + ' - expected distribution',
            modelout = modelout, plotout = plotout)

p = df.genPeakEst(3, counts, bins)
p['bg'].set(vary = False, value = 0)
p['mu0'].set(vary = False, value = 7)
p['sig0'].set(vary = False, value = 3.5)
p['A0'].set(vary= True, value = 500)
p['mu1'].set(vary = True, value = 0)
p['sig1'].set(vary = False, value = 3.5)
p['A1'].set(vary= True, value = 500)
p['mu2'].set(vary = False, value = 1.6)
p['sig2'].set(vary = False, value = 9.7)
p['A2'].set(vary = False, value = 106)
fitres5, AIC, *_, ll = df.fitNChidistr(p, bins, counts)
modelout = os.path.join(outdir, identifier + r'_expected+mid_distribution.txt')
plotout = os.path.join(outdir, identifier + r'_expected+mid_distribution.png')
df.plotdistr(distnorm, bins, fitres5, title = identifier + ' - expected + 1 free distribution',
            modelout = modelout, plotout = plotout)
print(fit_report(fitres5))
print('AIC for expected + 1 free distribution is %.1f' % AIC)
print('log-likelihood is ' + str(ll))


plt.plot(fstats['tauG'], fstats['proxRatio'], '.')
fitres
```
