```python
outdir = r'U:\Measurements\Singlem\singlem20-2\April\29_db_lowFRET_JHB\analysis_NV'
maxval = 60

LF29Aprs = df.filterFRETind(LF29Apr_an, 'stoichiometry', 0.2, 0.8)
#LF29Aprs = df.filterFRETind(LF29Aprs, 'sigmaY', 1.5, 2.5)

LF29Aprs = df.filterFRETind(LF29Aprs, 'NG', 50, 1000)
LF29Aprs = df.filterFRETind(LF29Aprs, 'NY', 100, 1000)

NF_stats = df.genStats(LF29Aprs)
#NF_stats = df.filterStats(NF_stats, 'sigmaY', 1.5, 2.5)
#NF_stats = df.filterStats(NF_stats, 'sigmaG', 0, 3.3)
plt.plot(NF_stats['distx'], NF_stats['disty'], 'o', alpha = 0.5)
plt.xlim(-30, 30)
plt.ylim(-30, 30)
plt.xlabel('dist x (nm)')
plt.ylabel('dist y (nm)')
plt.show()


NF_dist = np.array((NF_stats['distx'], NF_stats['disty'])).transpose()
#NF_dist = df.filterVec(NF_dist, maxval = maxval, center = [5,0])
NF_dist = df.filterVec(NF_dist, maxval = maxval)
NF_distnorm = np.linalg.norm(NF_dist, axis = 1)
outfile = os.path.join(outdir, r'NFRuler_distdict.txt')
distDict = {}
distDict ['distx'] = NF_dist[:,0]
distDict ['disty'] = NF_dist[:,1]
distDict['distnorm'] = NF_distnorm
df.saveDict(distDict, outfile)


modelout = os.path.join(outdir, r'Ruler_NF_chiFit.txt')
plotout = os.path.join(outdir, r'Ruler_NF_subpopulation.png')

TACout = os.path.join(outdir, r'TAC\NF_cut.txt')
df.subensembleTAC(LF29Apr_an, outfile = TACout)

dist = NF_distnorm
maxbin = maxval
binwidth = 1
counts, bin_edges, _ = plt.hist(dist, bins = np.arange(0, maxbin, binwidth))
plt.clf()
Nbins = bin_edges.shape[0] - 1
bins = np.zeros(Nbins)
for i in range(Nbins):
    bins[i] = (bin_edges[i] + bin_edges[i + 1]) / 2


fitres = df.whichChiIsBest(bins, counts, verbose = True)
df.plotdistr(NF_distnorm, bins, fit = fitres , title = 'LF Ruler', 
                           modelout = modelout, plotout = plotout)
p = df.genPeakEst(1, counts, bins)
p['bg'].set(vary = True, value = 0)
p['mu0'].set(vary = True, value = 8)
#p['mu0'].set(vary = False, value = 7.1)
#p['sig0'].set(vary = False, value = 4.6)
fitres2, AIC, *_ = df.fitNChidistr(p, bins, counts)
modelout = os.path.join(outdir, r'Ruler_NF_chiFit_mufree.txt')
plotout = os.path.join(outdir, r'Ruler_NF_subpopulation_mu0.png')
df.plotdistr(NF_distnorm, bins, fit = fitres2 , title = 'NF construct only', 
                           modelout = modelout, plotout = plotout)
p = df.genPeakEst(2, counts, bins)
p['bg'].set(vary = True, value = 0)
p['mu0'].set(vary = True, value = 14)
p['sig0'].set(vary = False, value = 3.5)
#p['A0'].set(vary = True, value = 150)
#p['mu1'].set(vary = False, value = 7)
#p['sig1'].set(vary = False, value = 2)
#p['mu2'].set(vary = True, value = 5)
#p['sig2'].set(vary = True, value = 12)
fitres3, AIC, *_ = df.fitNChidistr(p, bins, counts)
modelout = os.path.join(outdir, r'Ruler_NF_chiFit_mu0.txt')
plotout = os.path.join(outdir, r'Ruler_NF_subpopulation_mu0.png')
df.plotdistr(NF_distnorm, bins, fit = fitres3 , title = 'NF DNA Ruler - single-Chi', 
                           modelout = modelout, plotout = plotout)
outname = r'U:\Measurements\Singlem\singlem20-2\April\29_db_lowFRET_JHB\analysis_NV\single_Chi_components.csv'
df.exportChiComponents(outname, bins, fitres3.params)
print('AIC for mu fixed to 0 os %.1f' % AIC)
plt.plot(LF29Apr_stats['proxRatio'], LF29Apr_stats['tauG'], '.')
fitres
```
