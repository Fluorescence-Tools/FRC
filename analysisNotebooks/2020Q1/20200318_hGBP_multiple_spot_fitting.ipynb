```python
def plotfft(fft):
    """utility function"""
    #issue: fft_plot naming is non intuitive
    fft_plot = np.fft.fftshift(fft)
    fft_plot = np.log(np.abs(fft_plot))
    plt.imshow(fft_plot)
    plt.show()
def plotPairCorrelation(locLst, colorI, colorII, maxbin = 1000, pxSize = 10, binwidth = 20, ROIsize = 30, outfile = ''):
    """utility function"""
    dimx, dimy = locLst[0][color].bitmap.shape
    distances = imageWisePairCorrelation(locLst, colorI, colorII)
    ndist = len(distances)
    a = binwidth * 2 * np.pi / (dimx * dimy * pxSize ** 2) * ndist
    plt.figure(figsize = (7,5))
    counts, bins, *_ = plt.hist(np.array(distances)*pxSize, bins = np.arange(1,maxbin, binwidth), label = 'pairwise correlation, all data')
    plt.plot(np.arange(maxbin), np.arange(maxbin) * a, label = 'random distribution line')
    plt.plot([dimx * pxSize /4, dimx * pxSize /4], [0,max(counts)], 'r--', label = 'border effects become significant')
    plt.plot([ROIsize * pxSize , ROIsize * pxSize ], [0,max(counts)], 'b--', label = 'ROI size')
    plt.title('Pairwise correlation function, Yellow Channel')
    plt.xlabel('dye-to-dye distance (nm)')
    plt.ylabel('occurence / 10 nm')
    plt.legend()
    if outfile :
        plt.savefig(outfile, dpi = 300, bbox_inches = 'tight')
    plt.show

#depricated
#def pairDistance(posx, posy):
#    assert len(posx) == len(posy)
#    npairs = len(posx)
#    distances = []
#    for i in range(npairs):
#        for j in range(i+1, npairs):
#            distance = np.sqrt((posx[i] - posx[j])**2 + (posy[i] - posy[j])**2)
#            distances.append(distance)
#    return distances
def pairDistance(posxI, posyI, posxII, posyII):
    assert len(posxI) == len(posyI)
    assert len(posxII) == len(posyII)
    nI = len(posxI)
    nII = len(posxII)
    distances = []
    for i in range(nI):
        for j in range(nII):
            distance = np.sqrt((posxI[i] - posxII[j])**2 + (posyI[i] - posyII[j])**2)
            distances.append(distance)
    return distances

def imageWisePairCorrelation(locLst, colorI, colorII):
    distances = []
    for loc in locLst:
        posxI = []
        posyI = []
        posxII = []
        posyII = []
        for spot in loc[colorI].spotLst:
            posxI.append(spot.posx)
            posyI.append(spot.posy)
        for spot in loc[colorII].spotLst:
            posxII.append(spot.posx)
            posyII.append(spot.posy)
        distances += pairDistance(posxI, posyI, posxII, posyII)
    return distances

def addGauss(image, posx, posy, sigma):
    params = np.zeros(18)
    params[0] = posx
    params[1] = posy
    params[2] = 1 #amplitude
    params[3] = sigma
    params[4] = 1 #make symmetric Gauss
    params[5] = 0 #bg
    options = df.GAP.optionsCluster(setmodel = 0, ellipt_circ = 1, setbg = 0)
    options.transferOptions(params)
    image += GaussFits.model2DGaussian(params, image)
    return image
def drawLocImage(dimx, dimy, posx, posy, sigma):
    assert len(posx) == len(posy)
    npairs = len(posx)
    image = np.zeros((dimx, dimy))
    for i in range(npairs):
        image = addGauss(image, posx[i], posy[i], sigma)
    return image

def FourierRingSpectrum(fft):
    fft_shift = np.fft.fftshift(fft)
    xshape, yshape = fft.shape
    centerx = xshape // 2 #round down if shape is odd
    centery = yshape // 2
    return circularIntegral(centerx, centery, fft_shift)
def circularIntegral(centerx, centery, image):
    """calculates circular integral and puts it in a histogram.
    No. pixels in a bin is also returned"""
    dimx, dimy = image.shape
    #max distance is needed to initialize arrays
    deltax_max = max(centerx, dimx - centerx)
    deltay_max = max(centery, dimy - centery)
    dist_max = np.sqrt(deltax_max**2 + deltay_max**2)
    bins_max = np.ceil(dist_max).astype(np.int)+1
    bins = np.arange(bins_max)
    bincounts = np.zeros(bins_max)
    values = np.zeros(bins_max)
    
    for i in range(dimx):
        for j in range(dimy):
            dist = np.sqrt( (i - centerx)**2 + (j - centery) ** 2)
            binNumber = np.round(dist).astype(np.int)
            bincounts[binNumber] += 1
            #note that the complex values are added
            values[binNumber] += image[i,j]
    return bins, values, bincounts
```
