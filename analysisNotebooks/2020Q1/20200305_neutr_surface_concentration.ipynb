```python
import ImageManipulation as IM
import numpy as np
import matplotlib.pyplot as plt
import os
```

# run1

```python

def getbrightness(wdir):
    subdirs = os.listdir(wdir)
    for subdir in subdirs:
        files = os.listdir(os.path.join(wdir, subdir))
        intensity_tot = 0
        for i, file in enumerate(files):
            if file[-4:] == '.ptu':
                im =IM.processLifetimeImage(os.path.join(wdir, subdir, file).encode(), 
                                           uselines = np.array([1]))
                #plt.imshow(im.baseIntensity.R)
                #plt.colorbar
                #plt.show()
                intensity_tot += np.sum(im.baseIntensity.R)
        print('\n\tmeas: %s \t intensity %e\n' % (subdir, intensity_tot))
```

```python

def plotFolder(subdir):
    files = os.listdir(os.path.join(wdir, subdir))
    intensity_tot = 0
    for file in files:
        if file[-4:] == '.ptu':
            im =IM.processLifetimeImage(os.path.join(subdir, file).encode(), 
                                       uselines = np.array([1]))
            plt.figure(figsize = (6,6))
            plt.imshow(im.baseIntensity.R, cmap = 'hot')
            plt.colorbar
            plt.show()
```
