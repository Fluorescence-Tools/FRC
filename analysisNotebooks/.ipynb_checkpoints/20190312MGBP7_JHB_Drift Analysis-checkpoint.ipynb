```python
import FRCfuncs
import numpy as np
from os import path
import matplotlib.pyplot as plt
```

```python
wdir = r'N:\Singlem\singlem19-1\March\12_mGBP7-Project_JHB'
ext = r'DA_mGBP7-mGBP3_01\Green Photons'
Green = FRCfuncs.readfolder(path.join(wdir,ext))
```

```python
frames = Green.shape[0]
#first element of drift is 0
drift = np.zeros((frames,2))
for i in range(1, frames):
    drift[i], _ = FRCfuncs.DriftDetect(Green[i-1], Green[i], 200)
```

```python
drift*200
plt.rcParams["figure.figsize"] = [16,9]
totaldrift = np.cumsum(drift, axis = 0)
plt.plot(totaldrift[:,0], totaldrift[:,1], '.-')
```

```python
#windows = np.arange(100,0,-1)
windowsize = 100
length = drift.shape[0]
#only full windows are used.
nwindows = int(length / windowsize)
means = np.zeros(nwindows)
sigmas = np.zeros(nwindows)
for i in range(nwindows):
    window = drift[ i * windowsize : (i + 1) * windowsize]
    means[i] = np.mean(window)
    sigmas[i] = np.std(window)
```
