```python
def two2DGaussians(coord, A1, x1, y1, sigma, A2, x2, y2, offset):
    x = coord[0]
    y = coord[1]
#    xo = float(xo)
#    yo = float(yo)    
    g = offset + A1 * np.exp( - ((x-x1)**2 + (y-y1)**2) / (2 * sigma**2)) + \
        A2 * np.exp( - ((x-x2)**2 + (y-y2)**2) / (2 * sigma**2))
    return g.ravel()
```

```python
def findROIcenters(image, ROIfile, roiSide = 10, pxSize = 10, outdir = '', 
                   outname = '', showfig = True, savefig = False):
    """routine to fit two 2D Gaussians in snips of an image.
    image must be a 2D numpy array
    ROIfile contains the to be studied ROIs.
        The six entries are xcorner, xrange, ycorner, yrange, xcenter, ycenter
    roiSide is the side of the roi to be studied: xrange and yrange are ignored.
    pxSize is the pixel size of the original image in nm
    Outdir specifies the output folder.
    Function calls a plotting routine for visual inspection afterwards.
    Plotting function is called if either showfig or savefig is True.
    If showfig is true, output is printed in python dialog.
    If savefig is True, output is saved into outdir."""
    #load ROI file
    CLRROI = np.genfromtxt(ROIfile)
    
    #initialize arrays
    NROIs = CLRROI.shape[0] - 1 #last entry contains zeros
    outROI = np.zeros([NROIs*2, 6])
    j = 0
    #loop over all ROIs
    for i in range(NROIs): 
        #get ROI from image
        centery, centerx = CLRROI[i, [4,5]].astype(np.int)
        cornerx, cornery = CLRROI[i, [0,2]]
        xshape, yshape = image.shape[:2]
        if centerx - roiSide < 0 or centery-roiSide < 0 or centerx + roiSide > xshape or centery + roiSide > yshape:
            print ('ROI touches image border, skipping \n')
            continue
        else:
            ROIsnip = image[centerx - roiSide: centerx + roiSide, centery - roiSide: centery + roiSide]

        ### fit ROIsnip with two Gaussians
        # Create x and y indices
        x = np.arange(ROIsnip.shape[0])
        y = np.arange(ROIsnip.shape[1])
        coord = np.meshgrid(x, y)

        #execute fit
        try:
            popt, pcov = tryfit(coord, ROIsnip.ravel(), 0, p0 = (10, 10, 5, 4, 10, 10, 15, 1) )
        except AssertionError:
            print ('Fit did not converge for ROI %i after 10 tries, skipping' % i)
            continue
        p2p = np.sqrt((popt[1]-popt[5])**2 + (popt[2]-popt[6])**2) * pxSize
        print('distance between two peaks is %.0f nm' % p2p)

        if showfig or savefig:
            plotROI(ROIsnip, popt, showfig, savefig, outdir, outname, i)
            
        #add found ROI entries to output
        #empirical shift are applied to make AnI routine match python routine.
        #could be rounding errors, different counting ranges.
        outROI[j] = [np.floor(popt[1]) + cornerx - 3 -2,  
                        7, 
                        np.floor(popt[2]) + cornery - 3 -1,
                        7,
                        np.floor(popt[1]) + cornerx -2,
                        np.floor(popt[2]) + cornery -1] 
                        #xcorner, xrange, ycorner, yrange, xcenter, ycenter
        j += 1
        if p2p > 40:
            outROI[j] = [np.floor(popt[5]) + cornerx - 3 - 2, 
                        7, 
                        np.floor(popt[6]) + cornery - 3 -1, 
                        7,
                        np.floor(popt[5]) + cornerx -2,
                        np.floor(popt[6]) + cornery -1]
            j+= 1
        else:
            print ('ROI %i has two Gaussians closer than 40 nm, presumably single spot. \n' % i+
                'Skipping Second spot.' )
            
    #save output
    np.savetxt(os.path.join(outdir, outname), outROI, fmt = '%d', delimiter = '\t')
        
    return 0
```

```python
def tryfit(coord, ROIsnip, counter, p0):
    """recursive function that calls itself up to ten times or
        untill convergence is reached."""
    assert counter < 10,'after 10 tries, fit did not converge, skipping.'
    
    try:
        popt, pcov = curve_fit(two2DGaussians, coord, ROIsnip, p0 = p0)
    except RuntimeError:
        print ('Fit did not converge for ROI %i, trying again with new parameters' % i)
        counter += 1
        p0 = np.random.random(8) * np.array([20, 19, 19, 5, 20, 19, 19, 3])
        popt, pcov = tryfit(coord, ROIsnip, counter, p0)
    
    p2p = np.sqrt((popt[1]-popt[5])**2 + (popt[2]-popt[6])**2)
    
    if p2p < 3 or p2p > 10:
        print('distance between two peaks is less than 3 or more then ten pixels: retrying')
        counter += 1
        p0 = np.random.random(8) * np.array([20, 19, 19, 3, 20, 19, 19, 3]) + \
            np.array([0, 0, 0, 2, 0, 0, 0, 0])
        popt, pcov = tryfit(coord, ROIsnip, counter, p0)
        
    return popt, pcov
```

```python
def plotROI(ROIsnip, popt, showfig, savefig, outdir, outname, i):
    """routine specifically for plotting two Gaussian centers on an ROI
    sides of the box is three sigma.
    popt should be obtained from applying curve_fit to two2DGaussians
        to obtain the correct parameter ordering."""
    #draw image
    plt.imshow(ROIsnip)
    plt.colorbar()
    ax = plt.gca()
    
    #Draw two triangles to indicate ROIS
    x1, y1, w, h = [popt[1] - 1.5 * popt[3], popt[2] - 1.5 * popt[3], 3 * popt[3], 3 * popt[3]] #corner and sides of block
    x2, y2, w, h = [popt[5] - 1.5 * popt[3], popt[6] - 1.5 * popt[3], 3 * popt[3], 3 * popt[3]] #corner and sides of block
    rect1 = patches.Rectangle((x1, y1),w, h, linewidth=1,edgecolor='r',facecolor='none')
    rect2 = patches.Rectangle((x2, y2),w, h, linewidth=1,edgecolor='r',facecolor='none')

    ax.add_patch(rect1)
    ax.add_patch(rect2)

    #draw centers of Gaussians
    plt.scatter([popt[1], popt[5]], [popt[2], popt[6]], s = 3, c = 'r')
    
    #save and / or display image
    if savefig:
        try:
            os.mkdir (outdir)
        except FileExistsError:
            pass
        if outdir == '':
            raise ValueError ('No output dir to save figure specified, aborting.')
        outname = os.path.join( outdir, outname[:-4] + r'_ROI%i.png' % i)
        plt.savefig(outname, dpi = 300)
    if showfig:
        plt.show()
    return 0
```

```python
def sumdatfiles(wdir):
    os.chdir(wdir)
    files = os.listdir()

    for f in files:
        if f[-4:] == '.dat':
            try:
                tacPS += np.genfromtxt(f)
            except NameError:
                tacPS = np.genfromtxt(f)
    try: 
        os.mkdir('sum')
    except:
        pass
    np.savetxt(r'sum\sum.dat', tacPS, fmt = '%i')
    return tacPS
```
