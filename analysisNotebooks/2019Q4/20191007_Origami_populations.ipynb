```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def sumdatfiles(wdir):
    os.chdir(wdir)
    files = os.listdir()

    for f in files:
        if f[-4:] == '.dat':
            try:
                tacPS += np.genfromtxt(f)
            except NameError:
                tacPS = np.genfromtxt(f)
    try: 
        os.mkdir('sum')
    except:
        pass
    np.savetxt(r'sum\sum.dat', tacPS, fmt = '%i')
    return tacPS
```

```python
wdir = r'N:\Singlem\singlem19-4\october\07_origami_CF_NV\HF_560_10pct_640_005pct\HF_subpopulation\Total histograms_ROI and PN selected\Green Photons'
sumdatfiles(wdir)
wdir = r'N:\Singlem\singlem19-4\october\07_origami_CF_NV\NF_560_10pct_640_005pct\low_ensemble\Total histograms_ROI and PN selected\Green Photons'
sumdatfiles(wdir)
wdir = r'N:\Singlem\singlem19-4\october\07_origami_CF_NV\NF_560_10pct_640_005pct\mid_ensemble\Total histograms_ROI and PN selected\Green Photons'
sumdatfiles(wdir)
wdir = r'N:\Singlem\singlem19-4\october\07_origami_CF_NV\NF_560_10pct_640_005pct\high_ensemble\Total histograms_ROI and PN selected\Green Photons'
sumdatfiles(wdir)
wdir = r'N:\Singlem\singlem19-4\october\07_origami_CF_NV\LF_560_10pct_640_005pct\all_subensemble\Total histograms_ROI and PN selected\Green Photons'
sumdatfiles(wdir)
wdir = r'N:\Singlem\singlem19-4\october\07_origami_CF_NV\HF_560_10pct_640_005pct\NF_subpopulation\Total histograms_ROI and PN selected\Green Photons'
sumdatfiles(wdir)
```
