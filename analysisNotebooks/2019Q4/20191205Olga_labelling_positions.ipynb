```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import os
from scipy.signal import savgol_filter
```

```python
def Efficiency(r, R0 = 52):
    return 1 / (1 + (r/R0)**6)
```

```python
def pltddistr(RAB, linkerlength, pair):
    RAB_smooth = savgol_filter(RAB[1], 51, 3)
    weightedRAB = RAB_smooth / RAB_smooth.sum() * Efficiency(RAB[0])
    E = weightedRAB.sum()
    #kmean = E/(1-E)/2.4
    fig, axs = plt.subplots(2,1, sharex = True)
    fig.subplots_adjust(hspace=0)
    
    axs[0].tick_params(direction='in', top=True, right=True)
    axs[0].tick_params(direction='in', labelbottom=False)
    axs[0].plot(RAB[0], RAB[1]/ np.nansum(RAB[1]), label = 'raw')
    axs[0].plot(RAB[0], RAB_smooth/ RAB_smooth.sum(), label = 'smooth')
    axs[0].plot(RAB[0], weightedRAB, label = 'Efficiency weighted')
    axs[0].legend()
    axs[0].set_title('linker %i aa pair %s' % (linkerlength, pair))
    #axs[0].text(0,np.max(RAB_smooth/ RAB_smooth.sum())*.8, '<kfret> = %.2f 1/s\nE = %.2f' % (kmean, E))
    axs[0].text(0,np.max(RAB_smooth/ RAB_smooth.sum())*.8, '<E*> = %.2f' % E)
    rect1 = patches.Rectangle((0,0), 26, max(RAB_smooth), facecolor = 'gray')
    rect2 = patches.Rectangle((78,0), 272, max(RAB_smooth), facecolor = 'gray')
    axs[0].add_patch(rect1)
    axs[0].add_patch(rect2)
    plt.xlabel( u"R_DA distance (\u212B)")
    plt.ylabel("probability density function")
    
    axs[1].plot(RAB[0], Efficiency(RAB[0]), label = 'Efficiencies')
    axs[1].legend()
    
    #print(u'total E for linker length %i pair %s is %.2f' % (linkerlength, pair, E))
    return 0
```

```python
wdir = r'P:\CD95\sim\AV\results\linker51aa\distance_distributions'
files = os.listdir(wdir)
linkerlength = 51

for file in files:
    if file[-4:] != '.dat':
        continue
    pair = file.split('_histograms.dat')[0]
    RAB = np.genfromtxt(os.path.join(wdir, file))[:,1:]
    pltddistr(RAB, linkerlength, pair)
    plt.savefig(os.path.join(wdir, file[:-4] + '.png'), dpi = 300, bbox_inches = 'tight')
    plt.show()
```
