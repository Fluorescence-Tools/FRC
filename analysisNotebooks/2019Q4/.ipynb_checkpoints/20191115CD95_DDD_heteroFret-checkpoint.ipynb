```python
def batchAnalyseDA(files, outdir, D0_fname, dtime = 0.064):
    """utility function to batch-fit a set of Donor only calibrated
    Donor Acceptor fits. """
    #settings
    matplotlib.rcParams.update({'font.size': 16})

    #fit Donly
    D0_dat = np.genfromtxt(D0_fname)[30:600,1]
    popt_D0,_,Donly_base, Donlymodel, chi2red_D0 = fitDA.fitDonly(D0dat)

    names = []
    S_y = []
    x0 = []
    tau_fret = []
    chi2red_lst = []

    for file in files:
        #skip non-green files and Donly files
        if file[-6:] != '_G.dat' or file[:5] == 'Donly':
            continue

        names.append(file[:-6])
        #get brightness of Yellow signal
        S_y.append(np.sum(np.genfromtxt(os.path.join(wdir, file[:-6]+'_Y.dat'))[30:600,1]))
        #fit D(A)
        print('analyzing file %s' % file)
        DAdat = np.genfromtxt( os.path.join(wdir, file))[30:600,1]
        popt, pcov, DAmodel, chi2red = fitDA.fitDA(DAdat, D0dat, dtime)
        x0.append(popt[1])
        tau_fret.append(popt[2])
        chi2red_lst.append(chi2red)
        #plot and store
        print('Donly fraction is %.2f and FRET lifetime is %.2f ns' % (popt[1], popt[2]))
        fitDA.pltDA_eps(DAdat, D0dat, DAmodel, Donlymodel, file, popt, chi2red, chi2red_D0, outdir)
    return names, S_y, x0, tau_fret, chi2red_lst
```
