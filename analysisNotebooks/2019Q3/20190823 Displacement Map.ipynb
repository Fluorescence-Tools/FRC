```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
wdir = b'N:\\Singlem\\singlem19-3\\August\\23_JHB\\40by40um_correctionmap\\Colocalization Fit results'
fGreen = b'Green Photons_Fit results.0I4'
fRed = b'Red Photons_Fit results.II4'
px = 50 #pixelsize in nm
Gdat = np.genfromtxt(os.path.join(wdir, fGreen).decode(), skip_header = 1, delimiter = '\t')
Rdat = np.genfromtxt(os.path.join(wdir, fRed).decode(), skip_header = 1, delimiter = '\t')

#delete 'nan' entries
l = []
for i in range(Gdat.shape[0]):
    if ((not np.isnan(Gdat[i,2]) and (Gdat[i,1] != 0) and 
        (Gdat[i, 14]) < 50 ) and (Gdat[i,13] < 450 ) ):
        l.append(i)
Gdat = Gdat[l]
Rdat = Rdat[l]
dx = Gdat[:,2:4] - Rdat[:, 2:4]
pos = Gdat[:,2:4]
```

```python
plt.scatter(dx[:,0] * px, dx[:,1] * px, marker = '.')
plt.grid()
plt.xlim(-60, 60)
plt.ylim(-60, 60)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('displacement in x (nm)')
plt.ylabel('displacement in y (nm)')
plt.title('Channel displacement on D(0) sample')
plt.savefig('20190823ChannelDisplacementScatter40x40.png', dpi = 300, bbox_inches = 'tight')
print( 'displacement is %.1f nm in x and %.1fnm in y' %(dx[:,0].mean() * 50, dx[:,1].mean() * 50))
```

```python
plt.figure(figsize = (20,10))
plt.quiver(pos[:,0] * px , pos[:,1] * px , dx[:,0] * px, dx[:,1] * px )
plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
plt.xlabel('x field positions (nm)')
plt.ylabel('y field positions (nm)')
plt.savefig('20190823ChannelDisplacementVectorMap40x40.png', dpi = 300, bbox_inches = 'tight')
```
