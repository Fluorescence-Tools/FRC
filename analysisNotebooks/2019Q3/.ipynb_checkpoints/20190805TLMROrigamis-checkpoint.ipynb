```python
from PIL import Image
def saveGRYIm(RGYobj, outfolder, preposition = '', xmin = 0, xmax = -1, ymin = 0, ymax = -1): 
    #convert all strings to bytes
    try: outfolder = outfolder.encode()
    except: pass
    try: preposition = preposition.encode()
    except: pass
    try:
        os.mkdir(outfolder)
        print('creating new folder %s\n' %os.path.join(outfolder))
    except: 'folder already exists'
    im = Image.fromarray(RGYobj.G[xmin:xmax, ymin:ymax])
    outname = os.path.join(outfolder, preposition + b'imG.tif').decode()
    im.save(outname)
    im = Image.fromarray(RGYobj.R[xmin:xmax, ymin:ymax])
    outname = os.path.join(outfolder, preposition + b'imR.tif').decode()
    im.save(outname)
    im = Image.fromarray(RGYobj.Y[xmin:xmax, ymin:ymax])
    outname = os.path.join(outfolder, preposition + b'imY.tif').decode()
    im.save(outname)
    return 0
```

```python
import matplotlib.pyplot as plt
import numpy as np
from FRCfuncs import *
from scipy.optimize import curve_fit
import os
```

```python
def oneD_Gaussian(xdat, amplitude, xo, sigma):
    xo = float(xo)  
    g = amplitude*np.exp( - (xdat-xo)**2/(2*sigma**2))
    return g
```

```python
#Fit NF signal for origami 12. Data taken from imagej
ydat = [3.0938, 1.3210, 8.4139, 11.5000, 18.5815, 17.4907, 7.1701, 6.9228, 2.5417, 2.842E-14]
xdat = np.arange(len(ydat))*10 #10nm pixel size

popt1, pcov1 = curve_fit(oneD_Gaussian, xdat, ydat, p0 = (20, 50, 20) )
fit1 = oneD_Gaussian(np.arange(100), *popt1)

print(popt1[2]*2.35)
plt.plot(xdat, ydat, 'b')
plt.plot(np.arange(100), fit1, 'r--')
plt.plot([popt1[1]-popt1[2]*1.18,popt1[1]+popt1[2]*1.18],[popt1[0]/2,popt1[0]/2],'r-')
plt.text(40, 9.5, '%.0f' % (popt1[2]*2.35), fontsize=12)
plt.xlabel('distance (nm)')
plt.ylabel('intensity (cnts)')
plt.title('No-Fret Acceptor sensitized FWHM resolution')
plt.savefig('20190805linecut-NF.png', dpi = 300)
```
